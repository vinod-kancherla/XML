```python
from IPython import display
import xml.etree.ElementTree as ET
import urllib.request, urllib.parse, urllib.error

```

```python

data = '''
<person>
  <name>Vinod</name>
  <phone type="intl">
    +1 734 303 4456
  </phone>
  <email hide="yes" />
</person>'''

tree = ET.fromstring(data)
print('Name:', tree.find('name').text)
print('Attr:', tree.find('email').get('hide'))
```
